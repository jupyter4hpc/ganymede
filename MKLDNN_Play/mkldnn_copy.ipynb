```python
line = ''

def prune_mkldnn_verbose(logfile):
    key = 'mkldnn_verbose,'
    csvlines = []
    prunedlines = []
    with open(logfile,'r') as f:
        lines = f.readlines()
    for line in lines:
        if line.startswith(key):
            csvlines.append(line)
        else:
            prunedlines.append(line)
```
